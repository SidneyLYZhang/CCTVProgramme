# 解析示例

下面是一个整体解析的示例，使用针对性的解析包“dataparse”。

（1）导入解析包

```python
import dataparse as dpe
```

（2）确认数据文件夹

```python
filesplace = "./data/"
```

（3）获取现有数据的时间周期

```python
dateList = dpe.getDates(filesplace)
print("现有数据日期总长：{}天".format(len(dateList)))
dateList = sorted(dateList)
print("最早的日期：{}\n最晚的日期：{}".format(dateList[0],dateList[-1]))
```

（4）筛选某一天解析

```python
sel_date = "20240303"
```

```python
data = dpe.read_data(filesplace+"{}.json".format(sel_date))
```
